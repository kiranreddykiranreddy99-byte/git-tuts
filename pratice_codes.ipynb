```python
number = int(input('enter the number'));
if number < 0:
    print('This is a negative number.');
    if number % 2 == 0:
        print('This is an even number.');
    else:
        print('This is an odd number.');
elif number == 0:
    print('This is zero.');
else: # number > 0
    print('This is a positive number.');
    if number % 2 == 0:
        print('This is an even number.');
    else:
        print('This is an odd number.');
```
